```python
以台中市立委洪慈庸為例，依照月份來統計數值資料的變化
```

```python
import pandas as pd #引入環境內建函式庫pandas並簡寫其為pd
fb1=pd.read_csv('nysu_10902_2019立委_research.csv')#從電腦端資料夾中讀入csv數據庫並命名
fb2=pd.read_csv('nysu_10902_2019_research_right.csv')#讀入後來騰堯補給的資料
fb=pd.merge(fb1,fb2,how="outer")#將fb1和fb2取聯集
politics = pd.read_csv('9th_legislator_promise.csv')
```

```python
fb#叫出讀入的數據來看看
```

```python
hung = fb[fb.page_name == "洪慈庸"]#將有關洪慈庸粉專的資料庫從fb數據庫內抓出
hung.head(2)#
```

```python
hung[hung.created_time_taipei == max(hung.created_time_taipei)]#時間最晚的貼文
```

```python
hung[hung.created_time_taipei == min(hung.created_time_taipei)]#時間最早的貼文
```

```python
hung['new_date'] = pd.to_datetime(hung['created_time_taipei']).dt.date#將資料格created time轉為時間格式且最小單位是日期，並給與此筆資料new_date的名稱
hung['month_year'] = pd.to_datetime(hung['new_date']).dt.to_period('M')#將資料轉為時間格式且最小單位是月份
hung['year']=pd.to_datetime(hung['new_date']).dt.to_period('Y')#將資料轉為時間格式且最小單位是年
hung['post_hour'] = pd.to_datetime(hung['created_time_taipei']).dt.hour
hung.head(1)
```

```python
hung['new_date']
```

```python
hung['month_year']
```

```python
hung['year']
```

```python
#發文時間分佈
post_time=hung[['post_hour','message']].groupby('post_hour').count()
post_time.index.name = 'post_hour'
post_time.reset_index(inplace = True)
post_time
```

```python

import matplotlib.pyplot as plt 
labels = post_time['post_hour'].map(str)

fig,ax= plt.subplots()
fig.set_size_inches(15, 10)

ax.plot(labels, post_time['message'])

ax.legend()

plt.show()
```

```python
#整年貼文數量分佈
whole_year =hung[['message','month_year']].groupby('month_year').count()
whole_year.index.name = 'month_year'
whole_year.reset_index(inplace=True)
whole_year
```

```python
labels = whole_year['month_year'].map(str)

fig,ax= plt.subplots()
fig.set_size_inches(15, 10)

ax.plot(labels,whole_year['message'])

ax.legend()

plt.show()
```

```python
results = hung.groupby('month_year').sum()#將資料按年月分類並加總
results
```

```python
results.index.name = 'newhead' #將上行定義出的results索引標籤命名為newhead
results.reset_index(inplace=True)#將被重新命名過的索引標籤回復成系統本來的排列，個人覺得意義不大
results
```

```python
results['newhead'].map(str) #將索引標籤newhead那欄中所有資料轉為字串格式
```

```python
import matplotlib.pyplot as plt #引入環境內建matplotlib.pyplot繪圖函式庫

labels = results['newhead'].map(str) #將字串化的results['newhead']欄為引入為x軸
width = 0.5       # 柱狀圖中每根柱的寬度

fig, ax = plt.subplots()#在圖中分割出子圖
fig.set_size_inches(15, 8)#整張圖大小為15*8

ax.bar(labels, results['like_count'], width, label='like_count')#設定為柱狀圖
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()#顯示圖例

plt.show()
```
