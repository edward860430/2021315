```python
import pandas as pd 
fb = pd.read_csv('nysu_10902_2019立委_research.csv')
politics = pd.read_csv('9th_legislator_promise.csv')
```

以台中市立委洪慈庸為例，依照月份來統計數值資料的變化

```python
fb
```

```python
hung = fb[fb.page_name == "洪慈庸"]
hung.head(2)
```

```python
hung[hung.created_time_taipei == max(hung.created_time_taipei)]
```

```python
hung[hung.created_time_taipei == min(hung.created_time_taipei)]
```

```python
hung['new_date'] = pd.to_datetime(hung['created_time_taipei']).dt.date
hung['month_year'] = pd.to_datetime(hung['new_date']).dt.to_period('M')
hung.head(1)
```

```python
results = hung.groupby('month_year').sum()
results
```

```python
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
results['newhead'].map(str)
```

```python
import matplotlib.pyplot as plt

labels = results['newhead'].map(str)
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```
